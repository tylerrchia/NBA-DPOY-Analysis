Howevever, many fans and certain members of the media believe that Victor Wembanyama, the 7'4" elite rookie is more deserving of the DPOY award this year. We will take a look into who truly deserves the award based on past DPOY winners.

As of recent years, many argue that awards such as Most Valuable Player, Defensive Player of the Year, 6th Man of the Year should go to the best player in that category on the team with the or one of the best records. Others argue that Victor Wembanyama is clearly the best individual defensive player in the NBA this season, even though his team's defense does not reflect it.

For the purpose of the study, we will only consider the statistics starting from the 1983 season since that is when the DPOY award started.

## Column Names 

- **season**: season year
- **player**: player name
- **pos**: position
- **age**: age
- **tm**: team 
- **g**: games played during regular season
- **w**: games won durung regular season
- **mp**: total minutes played during the season
- **per** player efficiency rating (higher the better)
- **stl_percent**: percentage of defensive possessions that player got a steal
- **blk_percent**: percentage of defensive possessions that player got a block
- **drb_percent**: percentage of defensive possesions that player got a rebound
- **dws**: defensive win shares (higher the better)
    - Defensive Win Shares (DWS) is a basketball statistic that attempts to estimate the number of wins a player contributes to their team through their defensive performance. It is part of the Win Shares (WS) metric, which is widely used in basketball analytics to quantify a player's overall contribution to their team's success. DWS takes into account various defensive statistics such as blocks, steals, defensive rebounds, and individual defensive rating (points allowed per 100 possessions while the player is on the court). It also considers the player's playing time and the team's overall defensive performance. The formula for Defensive Win Shares is complex and proprietary, but it generally aims to attribute a player's defensive contributions to their team's success in terms of wins. A higher DWS indicates that a player has made a significant defensive impact on their team's performance, while a lower DWS suggests a lesser impact.
- **dbpm**: defensive box plus minus (higher the better)
    - Defensive Box Plus/Minus (DBPM) is a basketball statistic that aims to estimate a player's overall defensive impact on the game. It's a component of the Box Plus/Minus (BPM) metric, which is part of the advanced statistics used in basketball analytics. DBPM is calculated by attempting to measure a player's contribution to their team's defensive performance while they are on the court, relative to league average. It considers various defensive statistics such as blocks, steals, defensive rebounds, and personal fouls, along with the player's playing time and the team's performance while the player is on the court. A positive DBPM indicates that a player contributes positively to their team's defense, while a negative DBPM suggests that the player's defensive impact is below average. The scale is relative to average performance, so a DBPM of 0.0 represents league-average defense.
- **d_rtg**: defensive rating - measures the number of points a team allows per 100 possessions. It's a way to evaluate how effective a player or team is at preventing their opponents from scoring.
- **opp_e_fg_percent**: opposing team effective field goal percentage - statistic that quantifies the efficiency of a team's defense by considering the value of each field goal attempt made by their opponents. It accounts for the fact that three-pointers are worth more than two-pointers by giving them extra weight in the calculation. A lower eFG% for the opposing team signifies better defensive performance, indicating that the defense is effectively contesting shots and limiting the opponent's scoring efficiency. Teams with lower eFG% numbers typically force their opponents into more challenging shots, resulting in a decreased overall shooting percentage for the opposing team and reflecting strong defensive play.
- **pf_per_game**: personal fouls per game
- **stl_per_game**: steals per game
- **blk_per_game**: blocks per game
- **pf**: season total personal fouls
- **stl**: season total steals
- **blk**: season total blocks

## Individual Stats Exploration

We will be looking at the percent of steals and blocks each DPOY player had during their winning season. This helps standardize the data to show who truly is the most dominant defensive player since player minutes per game may vary.

```python
plt.figure(figsize=(10, 6))

# steal percent
plt.plot(past_advanced['season'], past_advanced['stl_percent'], color='blue', label='Steal Percent')
# block percent
plt.plot(past_advanced['season'], past_advanced['blk_percent'], color='red', label='Block Percent')

# wemby steal percent
plt.plot(advanced.loc[(advanced['player'] == 'Victor Wembanyama') & (advanced['season'] == 2024), 'season'], 
         advanced.loc[(advanced['player'] == 'Victor Wembanyama') & (advanced['season'] == 2024), 'stl_percent'], 
         color='orange', label='Wemby Steal Percent', marker='*', markersize = 10)

# wemby block percent
plt.plot(advanced.loc[(advanced['player'] == 'Victor Wembanyama') & (advanced['season'] == 2024), 'season'], 
         advanced.loc[(advanced['player'] == 'Victor Wembanyama') & (advanced['season'] == 2024), 'blk_percent'], 
         color='orange', label='Wemby Block Percent', marker='o', markersize = 8)

# wemby steal percent
plt.plot(advanced.loc[(advanced['player'] == 'Rudy Gobert') & (advanced['season'] == 2024), 'season'], 
        advanced.loc[(advanced['player'] == 'Rudy Gobert') & (advanced['season'] == 2024), 'stl_percent'], 
        color='indigo', label='Gobert Steal Percent', marker='*', markersize = 10)

# wemby block percent
plt.plot(advanced.loc[(advanced['player'] == 'Rudy Gobert') & (advanced['season'] == 2024), 'season'], 
        advanced.loc[(advanced['player'] == 'Rudy Gobert') & (advanced['season'] == 2024), 'blk_percent'], 
        color='indigo', label='Gobert Block Percent', marker='o', markersize = 8)

plt.xlabel('Season')
plt.ylabel('Percentages')
plt.title('Steal and Block Percent for DPOY Winners (Per 100 Possessions)')

plt.legend()
plt.grid(False)
# plt.savefig('stl_blk_%.png')
plt.show()
```

We see that Victor Wembanyama has the ***highest block percentage out of all past NBA DPOY winners*** as well as Rudy Gobert this season. ## Machine Learning Model

We will be creating a model to predict the DPOY of this year based on past DPOY winners. We will keep only the data from after 1983 since that is when the award was put into place.

```python
complete_advanced = pd.read_csv('Advanced.csv')

# Filter data for seasons after 1983
complete_advanced = complete_advanced[complete_advanced['season'] >= 1983]

# only keep needed columns
complete_advanced = complete_advanced[['season', 'player', 'pos', 'age', 'tm', 'g', 'mp', 'per', 'stl_percent', 'blk_percent', 'dws', 'dbpm', 'drb_percent', 'vorp']]

# Define conditions for DPOY winners
conditions = [
    ((complete_advanced['player'] == 'Jaren Jackson Jr.') & (complete_advanced['season'] == 2023)),
    ((complete_advanced['player'] == 'Marcus Smart') & (complete_advanced['season'] == 2022)),
    ((complete_advanced['player'] == 'Rudy Gobert') & (complete_advanced['season'] == 2021)),
    ((complete_advanced['player'] == 'Giannis Antetokounmpo') & (complete_advanced['season'] == 2020)),
    ((complete_advanced['player'] == 'Rudy Gobert') & (complete_advanced['season'] == 2019)),
    ((complete_advanced['player'] == 'Rudy Gobert') & (complete_advanced['season'] == 2018)),
    ((complete_advanced['player'] == 'Draymond Green') & (complete_advanced['season'] == 2017)),
    ((complete_advanced['player'] == 'Kawhi Leonard') & (complete_advanced['season'] == 2016)),
    ((complete_advanced['player'] == 'Kawhi Leonard') & (complete_advanced['season'] == 2015)),
    ((complete_advanced['player'] == 'Joakim Noah') & (complete_advanced['season'] == 2014)),
    ((complete_advanced['player'] == 'Marc Gasol') & (complete_advanced['season'] == 2013)),
    ((complete_advanced['player'] == 'Tyson Chandler') & (complete_advanced['season'] == 2012)),
    ((complete_advanced['player'] == 'Dwight Howard') & (complete_advanced['season'] == 2011)),
    ((complete_advanced['player'] == 'Dwight Howard') & (complete_advanced['season'] == 2010)),
    ((complete_advanced['player'] == 'Dwight Howard') & (complete_advanced['season'] == 2009)),
    ((complete_advanced['player'] == 'Kevin Garnett') & (complete_advanced['season'] == 2008)),
    ((complete_advanced['player'] == 'Marcus Camby') & (complete_advanced['season'] == 2007)),
    ((complete_advanced['player'] == 'Ben Wallace') & (complete_advanced['season'] == 2006)),
    ((complete_advanced['player'] == 'Ben Wallace') & (complete_advanced['season'] == 2005)),
    ((complete_advanced['player'] == 'Metta World Peace') & (complete_advanced['season'] == 2004)),
    ((complete_advanced['player'] == 'Ben Wallace') & (complete_advanced['season'] == 2003)),
    ((complete_advanced['player'] == 'Ben Wallace') & (complete_advanced['season'] == 2002)),
    ((complete_advanced['player'] == 'Dikembe Mutombo') & (complete_advanced['season'] == 2001) & (complete_advanced['tm'] == 'TOT')),
    ((complete_advanced['player'] == 'Alonzo Mourning') & (complete_advanced['season'] == 2000)),
    ((complete_advanced['player'] == 'Alonzo Mourning') & (complete_advanced['season'] == 1999)),
    ((complete_advanced['player'] == 'Dikembe Mutombo') & (complete_advanced['season'] == 1998)),
    ((complete_advanced['player'] == 'Dikembe Mutombo') & (complete_advanced['season'] == 1997)),
    ((complete_advanced['player'] == 'Gary Payton') & (complete_advanced['season'] == 1996)),
    ((complete_advanced['player'] == 'Dikembe Mutombo') & (complete_advanced['season'] == 1995)),
    ((complete_advanced['player'] == 'Hakeem Olajuwon') & (complete_advanced['season'] == 1994)),
    ((complete_advanced['player'] == 'Hakeem Olajuwon') & (complete_advanced['season'] == 1993)),
    ((complete_advanced['player'] == 'David Robinson') & (complete_advanced['season'] == 1992)),
    ((complete_advanced['player'] == 'Dennis Rodman') & (complete_advanced['season'] == 1991)),
    ((complete_advanced['player'] == 'Dennis Rodman') & (complete_advanced['season'] == 1990)),
    ((complete_advanced['player'] == 'Mark Eaton') & (complete_advanced['season'] == 1989)),
    ((complete_advanced['player'] == 'Michael Jordan') & (complete_advanced['season'] == 1988)),
    ((complete_advanced['player'] == 'Michael Cooper') & (complete_advanced['season'] == 1987)),
    ((complete_advanced['player'] == 'Alvin Robertson') & (complete_advanced['season'] == 1986)),
    ((complete_advanced['player'] == 'Mark Eaton') & (complete_advanced['season'] == 1985)),
    ((complete_advanced['player'] == 'Sidney Moncrief') & (complete_advanced['season'] == 1984)),
    ((complete_advanced['player'] == 'Sidney Moncrief') & (complete_advanced['season'] == 1983))
]

# binary variable to indicate if that player has won the DPOY or not
complete_advanced['DPOY_winner'] = np.select(conditions, [1] * len(conditions), default=0)
```

```python
from xgboost import XGBClassifier

# relevant features and target variable
X = complete_advanced[['stl_percent', 'blk_percent', 'drb_percent', 'dws']]
y = complete_advanced['DPOY_winner']

# Split data into training (previous seasons) and testing (2024 season)
X_train = X[complete_advanced['season'] < 2024]
y_train = y[complete_advanced['season'] < 2024]
X_test = X[complete_advanced['season'] == 2024]

# model and train it
model = XGBClassifier()
model.fit(X_train, y_train)

# predictions for the 2024 season
predictions = model.predict_proba(X_test)[:, 1]  # Probabilities of winning DPOY

# Find the player with the highest probability of winning
player_index = predictions.argmax()
player_name = complete_advanced.loc[complete_advanced['season'] == 2024, 'player'].iloc[player_index]

print(f"The player with the highest probability of winning the 2024 DPOY is: {player_name}")
```

```python
# index of Victor Wembanyama
victor_wembanyama_index = complete_advanced.loc[complete_advanced['season'] == 2024, 'player'].tolist().index("Victor Wembanyama")

# probability of Victor Wembanyama winning
victor_wembanyama_probability = predictions[victor_wembanyama_index]

print(f"The probability of Victor Wembanyama winning the 2024 DPOY award is: {victor_wembanyama_probability * 100} percent")


# index of Rudy Gobert
rudy_gobert_index = complete_advanced.loc[complete_advanced['season'] == 2024, 'player'].tolist().index("Rudy Gobert")

# probability of Rudy Gobert winning
rudy_gobert_probability = predictions[rudy_gobert_index]

print(f"The probability of Rudy Gobert winning the 2024 DPOY award is: {rudy_gobert_probability * 100} percent")
```

```python
# top 10 players with the highest probabilities of winning
top_players_indices = predictions.argsort()[-10:][::-1]
top_players = complete_advanced.loc[complete_advanced['season'] == 2024, 'player'].iloc[top_players_indices]
top_probabilities = predictions[top_players_indices]

# print the top 10 players and their probabilities
print("Top 10 Players for the 2024 DPOY Award:")
for player, probability in zip(top_players, top_probabilities):
    print(f"Player: {player}, Probability: {probability * 100} percent")
```

From the model, we see that Victor Wembanyama is the clear frontrunner for the Defensive Player of the Year Award for the 2024 season. According to the model which used steal percentage, block percentage, defensive rebound percentage, defensive win shares, and defensive box plus minus, Wemby has the highest probability of winning the Defensive Player of the Year Award at 7.12% according to the model. Rudy Gobert is not even in the top 10 of the model's predicted winners as he has a predicted probability of 0.039% of winning the award.
